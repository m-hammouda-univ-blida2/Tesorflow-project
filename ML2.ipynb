```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([
    keras.layers.Dense(4, input_shape=[1], activation='relu'),  # Couche cachée avec 4 neurones et activation ReLU
    keras.layers.Dense(1)  # Couche de sortie avec 1 neurone
])
#model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs=np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys=np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)

prediction = model.predict(np.array([[10.0]]))  # Input is now a 2D array

print(prediction)
```
